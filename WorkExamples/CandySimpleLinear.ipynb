```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

from scipy.stats import ttest_1samp  

################################### READ FILES ################################

CandySalesDF = pd.read_excel('Candy Sales Great chefs.xlsx',
                      skiprows=(0,1), header=None,
                      names=['OderDate', 'ProductID', 'ProductName','QuantitySold','UnitSalesPrice','TotalSalesRevenue',
                      'UnitCost','TotalCost','UnitProfit', 'TotalProfit', 'CustomerID', 'Customer', 'SalesRep'], usecols='A:M')
#################################################################################
# 1. Total sales revenue & Total Profit

Candy1xvar='TotalSalesRevenue'
Candy1yvar='TotalProfit'
Candy1x=CandySalesDF[Candy1xvar]
Candy1y=CandySalesDF[Candy1yvar]

Candy1slope, Candy1intercept, Candy1r_value, Candy1p_val, Candy1std_err = stats.linregress(y=Candy1y,x=Candy1x)
print("This is regression with Ho: X (total sales revenue) does not help to predict Y (total profit) /The slope is 0")

if np.sign(Candy1slope) < 0:   
    Candy1slsign = ""
else:
    Candy1slsign = "+"
    
Candy1regeq = f"{Candy1yvar} = {round(Candy1intercept,3)}{Candy1slsign}{round(Candy1slope,3)} {Candy1xvar}"

print(f"The equation is {Candy1regeq}")

print(f"The R-Squared is {round(Candy1r_value**2,4)} and the p-value is {round(Candy1p_val,4)}")

alpha=.05
if Candy1p_val < alpha:
    print("Conclusion: Reject Ho: X does help predict Y")
else:
    print("Conclusion: Fail to Reject Ho: We can't reject that X doesn't help to predict Y")

# Scatter plot using x (TV commercials) and y (Sales)
plt.scatter(Candy1x,Candy1y,color='black')
# Determine xy correlation
Candy1xyCorr = round(Candy1x.corr(Candy1y),3)
# Print subtitle with xy correlation, r-square, and p-value
plt.suptitle(f"Correlation: {Candy1xyCorr}  R-Squared: {round(Candy1r_value**2,4)} p-value: {round(Candy1p_val,4)}")
# Print title of regression equation
plt.title(Candy1regeq, size=10)
# Determine and print predicted y using red -
Candy1predict_y = Candy1intercept + Candy1slope * Candy1x
plt.plot(Candy1x,Candy1predict_y, 'r-')
plt.xlabel(Candy1xvar)
plt.ylabel(Candy1yvar)
plt.show()

#################################################################################
# 2. Unit Profit & Quantity Sold

Candy2xvar='UnitProfit'
Candy2yvar='QuantitySold'
Candy2x=CandySalesDF[Candy2xvar]
Candy2y=CandySalesDF[Candy2yvar]

Candy2slope, Candy2intercept, Candy2r_value, Candy2p_val, Candy2std_err = stats.linregress(y=Candy2y,x=Candy2x)
print("This is regression with Ho: X (unit profit) does not help to predict Y (quantity sold) /The slope is 0")

if np.sign(Candy2slope) < 0:   
    Candy2slsign = ""
else:
    Candy2slsign = "+"
    
Candy2regeq = f"{Candy2yvar} = {round(Candy2intercept,3)}{Candy2slsign}{round(Candy2slope,3)} {Candy2xvar}"

print(f"The equation is {Candy2regeq}")

print(f"The R-Squared is {round(Candy2r_value**2,4)} and the p-value is {round(Candy2p_val,4)}")

alpha=.05
if Candy2p_val < alpha:
    print("Conclusion: Reject Ho: X does help predict Y")
else:
    print("Conclusion: Fail to Reject Ho: We can't reject that X doesn't help to predict Y")

# Scatter plot using x (TV commercials) and y (Sales)
plt.scatter(Candy2x,Candy2y,color='black')
# Determine xy correlation
Candy2xyCorr = round(Candy2x.corr(Candy2y),3)
# Print subtitle with xy correlation, r-square, and p-value
plt.suptitle(f"Correlation: {Candy2xyCorr}  R-Squared: {round(Candy2r_value**2,4)} p-value: {round(Candy2p_val,4)}")
# Print title of regression equation
plt.title(Candy2regeq, size=10)
# Determine and print predicted y using red -
Candy2predict_y = Candy2intercept + Candy2slope * Candy2x
plt.plot(Candy2x,Candy2predict_y, 'r-')
plt.xlabel(Candy2xvar)
plt.ylabel(Candy2yvar)
plt.show()
```
